```python
import numpy as np
from matplotlib import pyplot as plt
```

# LinePlot

```python
x=np.arange(1,11)
```

```python
y=2*x
```

```python
plt.plot(x,y)
```

```python
plt.show
```

```python
plt.plot(x,y,color='g',linestyle=':',linewidth=2)
plt.title('Line Plot')
plt.xlabel('x_label')
plt.ylabel('y_label')
plt.show()
```

```python
x=np.arange(1,11)
y1=2*x
y2=3*x
```

```python
plt.plot(x,y1,color='r',linewidth=2,linestyle=':')
plt.plot(x,y2,color='g',linewidth=2,linestyle='-')
plt.title('Line plot b/w x,y1 and y2')
plt.xlabel('x_axis')
plt.ylabel('y_label')
plt.grid(True)
plt.show()
```

# subplot

```python
plt.subplot(2,1,1)
plt.plot(x,y1,color='r',linewidth=2,linestyle=':')

plt.subplot(2,1,2)
plt.plot(x,y2,color='g',linewidth=2,linestyle='-')
plt.title('Line plot b/w x and y2')
plt.xlabel('x_axis')
plt.ylabel('y_label')
plt.grid(True)
plt.show()      
```

# bar plot

```python
student={'naveed':30,'waseem':40,'asad':27}
```

```python
name=list(student.keys())
values=list(student.values())
```

```python
name
```

```python
values
```

```python
plt.bar(name,values)
plt.show()
```

```python
plt.bar(name,values,color='r')
plt.title('Bar Graph between name and values')
plt.xlabel('name')
plt.ylabel('values')
plt.show()
```

```python
plt.barh(name,values,color='g')
plt.title('Bar Graph between name and values')
plt.xlabel('name')
plt.ylabel('values')
plt.grid(True)
plt.show()
```

```python
x=[1,4,6,8,11,2,5,8,2,10]
y=[3,7,2,3,8,9,11,10,3,4]
yz=[2,6,7,3,11,9,4,6,3,9]
```

```python
plt.scatter(x,y,marker='*',c='g',s=100)
plt.scatter(x,yz,c='r',s=100)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.scatter(x,y,marker='*',c='g',s=100)
plt.title('X and Y')
plt.xlabel('x_label')
plt.ylabel('y_label')

plt.subplot(1,2,2)
plt.scatter(x,yz,c='r',s=100)
plt.title('X and yz')
plt.xlabel('x_label')
plt.ylabel('yz_label')
plt.show()
```
